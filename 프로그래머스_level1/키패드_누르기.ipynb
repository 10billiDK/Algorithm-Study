```python
# 왼손 엄지손가락은 *키패드에, 오른쪽 엄지손가락은 #키패드 위치에서 시작.
# 키패드 이동은 한칸만
# 1,4,7은 왼손 엄지손가락을 사용 / 3,6,9는 오른쪽 엄지손가락 사용
# 가운데 열의 2,5,8,0을 입력할 때는 두 엄지 손가락의 현재 키패드의 위치에서 더 가까운 엄지손가락을 사용
# 만약 두 엄지손가락의 거리가 같으면, 오른손잡이는 오른손 엄지손가락, 왼손잡이는 왼손 엄지손가락을 사용한다.

# 거리가 어느 정도인지 측정하려면 어떻게?
# 일단, 1,4,7/ 3,6,9조건 맞춰주기. 왼손 위치 리스트 필요. 오른손 위치 리스트 필요.   
def get_distance(hand,i):
    location ={'1':(0,0), '2':(0,1), '3':(0,2), '4':(1,0),'5':(1,1),'6':(1,2), '7':(2,0),'8':(2,1),'9':(2,2), '*':(3,0),'0':(3,1),'#':(3,2)}
    i = str(i)
    x_hand,y_hand = location[hand]
    x_i,y_i = location[i]
    return abs(x_hand-x_i)+abs(y_hand-y_i)

def solution(numbers, hand):
    answer = ''
    left='*'
    right='#'
    #number =[[1,2,3],[4,5,6],[7,8,9],[0]] 이중 리스트를 만들어서 원하는 수를 넣었을 때 인덱스 도출하는건 pandas를 써야 할걸..? iloc쓰고 싶은데..
    #그냥 딕트를 만들어서 좌표를 찾아 쓰게끔 하자.
    hand = 'R' if hand == 'right' else 'L'
    for i in numbers:
        if i ==1 or i ==4 or i==7:
            answer += 'L'
            left = str(i)
        if i==3 or i==6 or i==9:
            answer += 'R'
            right = str(i)
        if i==2 or i==5 or i==8 or i==0:
        
# 2,5,8,0의 경우에는 어떻게? 거리가 같으면 hands출력. 거리가 다르면 dict,abs를 활용한 거리재기
            
            dis1 = get_distance(left,i)
            dis2 = get_distance(right,i)
            
            if dis1>dis2:
                answer += 'R'
                right =str(i)
            elif dis1<dis2:
                answer += 'L'
                left =str(i)
            elif dis1==dis2:
                answer += hand
                if hand == "R":
                    right=str(i)
                elif hand =='L':
                    left=str(i)
    return answer

'''def get_distance(hand, number):
    location = {'1': (0, 0), '2': (0, 1), '3': (0, 2),
                '4': (1, 0), '5': (1, 1), '6': (1, 2),
                '7': (2, 0), '8': (2, 1), '9': (2, 2),
                '*': (3, 0), '0': (3, 1), '#': (3, 2)}
    number = str(number)
    x_hand, y_hand = location[hand]
    x_number, y_number = location[number]
    return abs(x_hand - x_number) + abs(y_hand - y_number)

def solution(numbers, hand):
    answer = ''
    left, right = '*', '#'
    hand = 'R' if hand == 'right' else 'L'
    for number in numbers:
        if number in [1, 4, 7]:
            answer += 'L'
            left = str(number)
            continue
        if number in [3, 6, 9]:
            answer += 'R'
            right = str(number)
            continue
        if number in [2, 5, 8, 0]:
            dis1 = get_distance(left, number)
            dis2 = get_distance(right, number)
            if dis1 > dis2:
                answer += 'R'
                right = str(number)
            if dis1 < dis2:
                answer += 'L'
                left = str(number)
            if dis1 == dis2:
                answer += hand
                if hand == 'R':
                    right = str(number)
                if hand == 'L':
                    left = str(number)
    return answer'''
```

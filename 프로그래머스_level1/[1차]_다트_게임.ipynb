```python
# 1. pass : 실행할 코드가 없는 것으로 다음 행동을 계속해서 진행합니다.
# 2. continue : 바로 다음 순번의 loop를 수행합니다.
# 3. break : 반복문을 멈추고 loop 밖으로 나가도록합니다.
# dictionary --> if, elif문으로 나눠서 써야하는 상황 속에서 key값을 넣어서 item을 찾아쓰게끔 하면 if,elif문을 쓰는 수고를 덜 수 있다.

# 원래는 0,10을 고려하지 못했었음. 추가적으로 싱글, 더블, 트리플 타입 구분과 옵션을 구분 하는것에 있어서 문어발식으로 문제를 해결하려 함. --> 생각을 이렇게 할 필요가 없었다. 그냥 for문 돌려서 문자 나오면 어떻게 해라, 옵션 나오면 어떻게 해라 이렇게 쉽게 가도 됐음. --> 근데 이것도 딕셔너리 만들어서 생각하면 더 깔끔하게 해결할 수 있었다.

# 가장 중요하게 생각해야할게 무수한 조건들을 먼저 최대한 생각해놓고 풀기.

# 분기를 나눠서 각 분기 값의 누적합으로 전체 합을 구하려고 했지만, 나중에 나오는 *나 #에 따라 앞에 나오는 값의 계산값이 바뀌기 때문에 리스트에 각 세 분기 각각의 합을 따로 구하고 마지막에 더했다. flag는 분기 기점이자 a리스트의 인덱스.

'''def solution(dartResult):
    bonus = {'S':1, 'D':2, 'T':3}
    option = {'*':2, '#':-1}
    
    a= [0,0,0]
    flag=-1
    
    for idx, dart in enumerate(dartResult):
        if dart.isdigit():
            flag +=1
            if dart =='0':
                continue
            elif dartResult[idx+1].isdigit():
                a[flag] = int(dart)*10
                flag -=1
            else:
                a[flag] = int(dart)
        elif dart in 'SDT':
            a[flag] **=bonus[dart]
        else:
            if dart =="*":
                a[flag-1] *=2
            a[flag] *= option[dart]
            
    return sum(a)'''

#another solution
def solution(dartResult):
    bb=list(dartResult)
    answer=[]
    d=[]
    
    for i in range(len(bb)):
        if bb[i] =='1' and bb[i+1]=='0':
            d.append('10')
        elif bb[i] =='0' and bb[i-1]=='1':
            continue
        else:
            d.append(bb[i])
    for i in range(1,len(d)):
        if d[i] =='S':
            answer.append(int(d[i-1]))
        elif d[i] =='D':
            answer.append(int(d[i-1])**2)
        elif d[i] =='T':
            answer.append(int(d[i-1])**3)
        
        if d[i] =='*':
            if len(answer)>=2:
                answer[-1]=answer[-1]*2
                answer[-2]=answer[-2]*2
            else:
                answer[-1]=answer[-1]*2
        elif d[i]=='#':
            answer[-1]=answer[-1]*(-1)
    return sum(answer)
```
